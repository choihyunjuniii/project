```python
import matplotlib.pyplot as plt
```

### 1.선그래프

```python
y=[2,4,6]

plt.plot(y)   #그래프그려주는 역할  y변수에담긴것그림
plt.show()
```

```python
y=[2,4,6]
x=[1,2,3]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linestyle='--')
plt.show()
```

```python
#해당하는 점확인
plt.plot(y,marker='o')
plt.show()
```

```python
#,굵게
plt.plot(x,y,linewidth=5)
plt.show()
```

```python
#색깔
plt.plot(x,y,color='r')
plt.show()
```

```python
#연습문제
x=[1,2,3,4,5,6,7]
y=[1,4,5,8,9,5,3]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,marker='o')
plt.show()
```

```python
plt.plot(x,y,linestyle=':')
plt.show()
```

```python
plt.plot(x,y,linestyle='-.')
plt.show()
```

```python
plt.plot(x,y,linestyle=':',marker='D',markersize='10',color='g')
plt.show()
```

```python
plt.plot(x,y,ls='-',c='c')
plt.show()
```

```python
plt.plot(x,y,ls='-',c='c',marker='H',markersize='15')
plt.show()
```
