```python
import os
from tensorflow.keras.preprocessing.image import load_img
```

```python
base_folder = './google_img'
result_list = os.listdir(base_folder)
```

```python
result_list
```

```python
for dir_name in result_list :
    print(base_folder+'/'+dir_name)
```

```python
for dir_name in result_list :
    #경로를 결합할때 사용하는 함수
    category_path = os.path.join(base_folder, dir_name)
    print(category_path)
```

```python
for dir_name in result_list :
    #경로를 결합할때 사용하는 함수
    category_path = os.path.join(base_folder, dir_name)
    fnames = os.listdir(category_path)  #각 폴더별 파일 이름 얻기
    print(fnames)
```

```python
for dir_name in result_list :
    #경로를 결합할때 사용하는 함수
    category_path = os.path.join(base_folder, dir_name)
    fnames = os.listdir(category_path)  #각 폴더별 파일 이름 얻기
    for file in fnames :
        img_path = os.path.join(category_path, file)
        print(img_path)
```
