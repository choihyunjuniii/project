```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.show()
```

```python
#범위 지정 (그림의 영역 지정)

x = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
#xlim : x범위 지정
#ylim : y범위 지정
plt.ylim(-4,12)
plt.xlim(-2,8)

plt.show()
```

```python
#tick 설정
#xticks : x축 틱 설정
#yticks : y축 틱 설정

plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15
        , mec = 'g', mew = 5, mfc = 'r')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
#grid
plt.plot(x,y,ls = '--', lw = 5 , marker = 'o', ms = 15 , mec = 'g', mew = 5, mfc = 'r')
plt.grid()
plt.show()
```

```python
temp=data['요일'].value_counts()
temp
```

```python
temp['화']
```

```python
#월화보고싶을때 리스트로 묶어라
temp[['월','화']]
```

```python
y=temp[['월','화','수','목','금','토','일']]
x=y.index
```

```python
y.index
```

```python
#bar chart, x축에 대해 값 지정 X
plt.bar(x,y)
plt.title('요일별 사고량')
plt.xlabel('요일')
plt.ylabel('사고건수',rotation=0)

plt.show()
```

```python
#차이를 잘 보이기 위해서
plt.bar(x,y)
plt.ylim(500,620)  #y축의 범위
plt.show()
```

```python
pd.set_option('display.max_colum',30)  #30개의 컬럼값을 보여줘
```

```python
data
```

```python
car_to_car=data[data['사고유형_대분류']=='차대차']  #사고유형 대분류 컬럼에 차대차사고갯수
car_to_car
```

```python
#groupby()- 같은 값을하나로 욲어 통계 또는 집계 결과를 얻기 위해 사용하는것
car_to_car_place=car_to_car[['사상자수','발생지시도']].groupby('발생지시도').sum() #썸대신 mean쓰면 평균값나올것
car_to_car_place
```

```python
car_to_car_place.index  #인덱스값
```

```python
car_to_car_place.values  #발류값
```

```python
x=car_to_car_place.index
```

```python
y=car_to_car_place.values
```

```python
plt.figure(figsize=(15,8))  #이건 맨위에 써줘야함 그래프가 크게 쟉게 만듬
plt.plot(x,y)
plt.xlabel('도시')
plt.show()
```
