```python
#Request 사용법
#페이지 정보를 요청할때 사용하느 라이브러리
```

```python
import requests as req
```

```python
#네이버 페이지를 요청
#요청을 도와주는 라이브러리 = req이다.
res = req.get("https://www.naver.com/")
```

```python
#Response [200] = 페이지를 잘 받아왔습니다!
```

```python
#요청한 페이지의 정보를 보기!
res.text
```

```python
#사람으로 속이는 작업
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36"}
```

```python
mel = req.get("https://www.melon.com/", headers = h)
```

```python
mel
```

```python
#Beautifulsoup 사용하기
#--가져온 데이터에서 내가 필요한 정보들만 주출해주는 라이브러리
```

```python
from bs4 import BeautifulSoup as bs
```

```python
#bs(어떤 데잍를 처리할건지, 어떻게 처리 = lxml)
soup = bs(res.text, 'lxml')
```

```python
soup
```

```python
#내가 가지고온 데이터에서 A태그중에 클래스명 nav만 가지고와줘
result = soup.find_all('a', class_ = 'nav')
```

```python
#가지고온 데이터에서 순수하게 글자만 추출할때는 .text기능 활용
result[2].text
```

```python
result[0].text
```

```python
#리스트의 전체 데이터를 조회(글자추출)
```

```python
for i in result : 
    print(i.text)
```
