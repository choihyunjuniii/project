```python
#과목별 가장 근 값과 작은 값의 차이
maxArr=score.loc[:'Web',:'4반'].max(axis=1)  # 기본은 axis=0 제가 앞에서 다 같이 설명 드릴게요!
maxArr
```

```python
#apply변환
#행이나 열 단위로 더 복잡한 처리르 하고 싶을 때 사용!
#df명.apply(커스텀함수=내가만든함수,axis=?)
```

```python
#과목별 max점수와 min점수 차이 구하는 함수
def max_min(x):
    return x.max()-x.min()    #시프트엔터하면 함수생성
```

```python
#과목별 최대최소차이 값 쉽게 구할 수 있다.
score.loc[:'Web', :'4반'].apply(max_min, axis=1)
```

#### 카테고리화

```python
#케글의 타이타닉-나이범위를 group화 했던것처럼 0~10ㄷ과 10~20
```

```python
ages=[0,2,10,21,23,37,31,61,20,41,32,100]  #나이  >>0과100은 범위밖이여서 NaN나온것
#카테고리로 변경시 어떻게 구간을 지정할건지
#1~15(미성년자), 16~25(청년), 26~35(중년), 36~60(장년), 61~99(노년)
bins=[0,15,25,35,60,99]  #그에해당하는 구간

labels=['미성년자','청년','중년', '장년','노년']  #구간의 이름
#원하는 구간을 잘라주는 cut() 카테고리화 함수
cats=pd.cut(ages,bins,labels=labels)
cats
```

```python
#Length라는 범위?갯수
```

```python
type(cats)
```

```python
#카테고리 확인
cats.categories
```

```python
#카테고리화한 데이터를 보기쉽게 표로 만들어보기
age=pd.DataFrame(ages, columns=['나이'])
age
```

### DataFrame 병합
두가지 함수가있음
-concat
-merge 

```python
df1=pd.DataFrame({'A':['AO','A1','A2','A3'],
             'B':['B0','B1', 'B2','B3'],
             'C':['C0','C1','C2','C3'],
             'D':['D0','D1','D2','D3']},index=[0,1,2,3])
df1
```

```python
df2=pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},index=[4,5,6,7])
df2
```

```python
df3=pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},index=[8,9,10,11])
df3
```

```python
#concat axis=0 방향
result1=pd.concat([df1,df2,df3])
result1
```

```python
#concat axis=1방향
#알수 없는 데이 터는 결측값 반환
result2=pd.concat([df1,df2,df3],axis=1)
result2
```

```python
#keys 속성을 이용해서 df 구분
# 다중 index가 된다.
result=pd.concat([df1,df2,df3],keys=['x','y','z'])
result
```

```python
result.index
```

```python
df4=pd.DataFrame({'B':['B2','B3','B6','B7'],
                 'D':['D2','D3','D6','D7'],
                 'F':['F2','F3','F6','F7']},
                 index=[2,3,6,7])
df4
```

```python
#axis=0방향으로 병합
pd.concat([df1,df4])
```

```python
#axis=1방향으로 병합
pd.concat([df1,df4],axis=1)
```

```python
#공통으로 있는것만 가져오기
#join 속성사용
#inner : 교집합, 합치는 데이터들이 동일하게 가지고 있는 기준만 출력
#outer : 합집합, 기본값
#기본값 join='inner' or 'outer'
pd.concat([df1,df4],axis=1, join='inner')
```

```python
pd.concat([df1,df4],axis=1, join='outer')
```

```python
pd.concat([df1,df4],axis=0, join='inner')
```

```python
#인덱스를 무시하고 다시 부여해주세요 
#ignore_index >> 기존에 가지고 잇던 인덱스를 무시하고 새로운 인덱스 번호부여
#ignore_index=True(False 기본값)
pd.concat([df1,df4],ignore_index=True)
```

```python
#merge
#공통된 데이터를 기중으로 데이터를 합침
```

```python
#데이터 생성
df5=pd.DataFrame({'key':['K0','K2','K3','K4'],
                 'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],})
df5
```

```python
df6=pd.DataFrame({'key':['K0','K1','K2','K3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['K0','D1','D2','D3'],})
df6
```

```python
#공통된 컬럼('key')를 기준으로 합쳐준다
pd.merge(df5,df6,on='key')
```

```python
#how
#병합의 방식 지정
#concat의 join과 동일한 기능
#outer : 합집합, 전체 값 출력
#inner : 교집합, 공통된 데이터의 값만 출력
```

```python
#how='inner'가 기본값   concat과는 반대
pd.merge(df5,df6,how='inner')  #공통된것
```

```python
pd.merge(df5,df6,how='outer')
```

```python
#how='right', 오른쪽 df를 'key기준으로 출력
pd.merge(df5,df6,how='right')  #df6가 기준
```

```python
df_merge=pd.merge(df5,df6,how='left') #df5가 기준
df_merge
```

```python
df_merge.fillna(0)  #NaN  >> 0 채우기
```
